```python
import json
import pandas as pd
import os
import sys
import json
import openml
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# random_forest, AdaBoost, SVM, ExtraTrees, GradientBoosting, DecisionTree
# cls = "AdaBoost"
for cls in ["random_forest", "AdaBoost", "SVM", "ExtraTrees", "GradientBoosting", "DecisionTree"]:
    with open(f'../output/json/modified_losses_datasets_{cls}.json') as json_file:
        modified_losses_dataset = json.load(json_file)

    with open(f'../output/json/uniform_losses_datasets_{cls}.json') as json_file:
        uniform_losses_dataset = json.load(json_file)
    
    losses_differences = {key: uniform_losses_dataset[key] - modified_losses_dataset.get(key, 0) for key in uniform_losses_dataset}
    df_losses_differences =  pd.DataFrame.from_dict(losses_differences, orient = 'index')

    # Create the boxplot
    plt.figure(figsize=(5, 15))
    plt.violinplot(df_losses_differences, showmeans= True)
    plt.tick_params(axis='x', which='both', bottom='off', top='off', labelbottom='off')
    plt.plot([0.5, 1.5], [0, 0], 'k-', linestyle='--', lw=1)
    ax = plt.gca()
    ax.axes.xaxis.set_visible(False)
    plt.grid(False)
    plt.savefig(f'../output/verified_priors/verified_proirs_{cls}.pdf')
    plt.show()
    print("Done!")
```
