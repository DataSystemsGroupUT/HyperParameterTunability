```python
import math
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python

def plot_timecomplexity_basedon_hyper_value(cls,hyperparameter,time_complexity):
    full_path='../PerformanceData/'+ cls+"_results_total.csv"
    df = pd.read_csv(full_path)
    df=df[[hyperparameter,time_complexity]]
    grouped=df.groupby(hyperparameter)[time_complexity].mean().reset_index()
    plt.plot(grouped[hyperparameter],grouped[time_complexity]);
    plt.title(cls);
    plt.xlabel(hyperparameter);
    plt.ylabel(time_complexity);
    pathsave='../output_plots/'+cls+hyperparameter+time_complexity+'.png'
    plt.savefig(pathsave)
    return True
```

```python
plot_timecomplexity_basedon_hyper_value("RF","min_samples_leaf",'Mean_Train_time');
```
